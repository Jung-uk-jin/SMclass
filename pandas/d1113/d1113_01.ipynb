```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
#한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 글자크기
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
import pandas as pd
df = pd.read_excel('score.xlsx')
df
```

```python
plt.plot(df.index,df['국어'],label="국어")
plt.plot(df.index,df['영어'],label="국어")
plt.plot(df.index,df['수학'],label="국어")
```

```python
def func(x):
  if x-30<0:
    return 0
  else:
    return x-15
plt.bar(df['이름'],df['국어'],width=0.6,label="국어")
plt.bar(df['이름'],df['영어'],bottom=df['국어'],width=0.6,label="영어")
plt.bar(df['이름'],df['수학'],bottom=df['국어']+df['영어'],width=0.6,label="수학")
plt.legend()
n_list = df['국어'].apply(func)
for i in range(8):
  plt.text(df['이름'][i],n_list[i],df['국어'][i],ha="center")
  plt.text(df['이름'][i],df['국어'][i]+df['영어'][i]-15,df['영어'][i],ha="center")
  plt.text(df['이름'][i],df['국어'][i]+df['영어'][i]+df['수학'][i]-10,df['영어'][i],ha="center")
plt.show()
```

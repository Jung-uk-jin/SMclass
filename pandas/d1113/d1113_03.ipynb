```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
#한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 글자크기
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
import pandas as pd
df = pd.read_excel('score.xlsx',index_col="지원번호")
df
```

```python
values = [1,1,1,1,2,1]
explode=[0.1,0.1,0,0,0,0]
labels = ['python','java','javascript','oracle','html','etc']
colors = ['plum','beige','skyblue','b','r','c']
plt.title("프로그램 사용률")
#labels : 제목 explode 벌어지는 간격, counterclock False : 시계 반대방향, startangle : 시작하는 각도, autopct : 퍼센트 공개
plt.pie(values,labels=labels,explode=explode,autopct='%.1f%%',counterclock=False,startangle=90,colors=colors)
plt.legend(loc=(1.2,0.3))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
#한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 글자크기
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
import pandas as pd
df = pd.read_excel('score.xlsx')
```

```python
index = np.arange(df.shape[0])
index
```

```python
w=0.25
plt.bar(index,df['국어'],width=w,label='국어')
plt.bar(index-w,df['영어'],width=w,label='영어')
plt.bar(index+w,df['수학'],width=w,label='수학')
plt.xticks(index)
for i in range(8):
  plt.text(index[i]-0.2,df['국어'][i],df['국어'][i])
  plt.text(index[i]-0.5,df['영어'][i],df['영어'][i])
  plt.text(index[i],df['수학'][i],df['수학'][i])
plt.legend(ncol=3)
plt.show()
```

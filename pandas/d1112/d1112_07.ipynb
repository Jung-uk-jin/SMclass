```python
import matplotlib.pyplot as plt
import matplotlib
#한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 글자크기
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
import pandas as pd
```

```python
x=[1,2,3]
y=[2,4,8]
y2=[5,1,3]
y3=[1,2,5]
plt.figure(figsize=(10,5))
plt.plot(x,y,label="SM매출액",marker='o')
plt.plot(x,y2,label="LG매출액",marker='o')
plt.plot(x,y3,label="SK매출액",marker='o')
for i in range(3):
  plt.text(x[i],y[i]-0.5,y[i],ha='center')
  plt.text(x[i],y2[i]-0.5,y2[i],ha='center')
  plt.text(x[i],y3[i]-0.5,y3[i],ha='center')
plt.title("기업별 매출액")
plt.legend(ncol=3) # 기본 : 세로출력, 세로출력 : ncol=3
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib
#한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 글자크기
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
import pandas as pd
import random
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
x=[1,2,3,4,5,6,7,8]
x=df.index
y= df['국어']
yy=df['수학']
z=df['영어']
plt.figure(figsize=(20,5)) #그래프 크기 설정 x축으로 5 y축으로 20 dpi : 그래픽 해상도 값이 높을수록 크기 커짐
plt.plot(x,y,label="국어",marker='o',ms=8,ls='--')
plt.plot(x,yy,label="수학",marker='v',ms=8)
plt.plot(x,z,label="영어",marker='o',ms=8)
plt.legend() # 범례 생성
plt.show()
# plt.savefig('graph.png',dpi=300) # 파일저장
```

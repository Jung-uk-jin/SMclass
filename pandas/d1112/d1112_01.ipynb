```python
import pandas as pd
import random
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df['요일']=0 # 컬럼 추가
df
```

```python

df.loc['1번','요일']=5  # df.loc => row부분 수정 [x,y] x row축에 있는 y컬럼 값 
df.iloc[0,9] #iloc [x,y] row인덱스 x번에 있는 컬럼 인덱스 y번에 있는 값
```

```python
#랜덤으로 특정 셀에 0~6사이에 랜덤숫자 입력
for i in range(8):
  df.iloc[i,9] = random.randint(0,6)
df
```

```python
# 함수 ㅓ\정의
def func(x):
  if x==0: return '일요일'
  elif x==1: return "월요일"
  elif x==2: return "화요일"
  elif x==3: return "수요일"
  elif x==4: return "목요일"
  elif x==5: return "금요일"
  elif x==6: return "토요일"
```

```python
df['요일'].apply(func)    
```

```python
df['요일'] = df['요일'].apply(func)
df
```

```python
df['결과'] = '불합격'
df

## 국어 점수가 60이상이면 합격
```

```python
filt = df['국어'] > 60
df.loc[filt,'결과'] = '합격'
df
```

```python
#학교 + 등학교
# 키 + 'cm'
```

```python
df['키'] = df['키'].astype(str)+ 'cm'
df
```

```python

df.groupby('학교')['이름'].count() # 학교마다 이름 개수
df.groupby('학교')['국어'].mean() # 학교마다 국어 평균
df.groupby('학교')['국어'].sum() # 학교마다 국어 합계
df.groupby('학교')['국어'].sum().sort_values(ascending=False) # 학교마다 국어 합계 정렬
```

컬럼이 숫자가 아니면 실행이 안됨, mean(),max(),min(),std()

```python
df.groupby(['학교'])['국어'].agg(['mean','min','max','sum']) # 한번에 보기
```

```python
df['학년'] = [3,3,2,1,1,3,3,2]
df.groupby(['학교','학년'])['국어'].agg(['mean','min','max','sum']) # 학교와 학년을 나눠 출력
```

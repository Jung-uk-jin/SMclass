```python
import matplotlib.pyplot as plt
import matplotlib
#한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 글자크기
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
import pandas as pd
```

```python
x=[1,2,3] # 리스트 튜플 Series 등 형태
y=[2,4,8]
yy=[1,9,2]
   # x=[1,2,3,4,5,6,7,8]
# x=df.index
# y=df['수학']
plt.plot(x,y,label = '인원증가',linewidth = 5,marker='o',ms=15,mfc='green',ls ='--') # 그래프 생성, label 범례 제목, linewidth : 선 두께
# 그래프 생성, label 범례 제목, marker : 지점 표시- o,x,v ,markersize = ms : 마커 크키, markerfacecolor = mfc : 마커 색 변경
# alpha : 투명도 설정 ls : 라인스타일 : 점선 -- dash선, color 선 색상
plt.plot(x,yy,label = '비용증가',color='pink',marker='v',ms=15,mfc = 'yellow',alpha = 0.5, ls=':') 
plt.title("라인그래프LIne",size=30) # 그래프 제목 , size 글자크기 변경

plt.xlabel('x축 날짜',color = 'red',loc = 'right') # x축 설명 , color= 글자색 변경, loc = 글자위치 변경 left,right,center
plt.ylabel('y축 날짜', color = '#8E3422',loc='top') # y축 설명 loc = 글자위치 변경 top,center,bottom
plt.xticks([1,2,3]) # x축 눈금 크기 설정
plt.yticks([2,4,8]) # y축 눈금 크기 변경
#범례 : 그래프 선 설명 문구, loc:  범례위치 upper, lower, left,center,right
# plt.legend(loc='upper center')
plt.legend(loc=(0.1,0.8)) # 좌표값으로 위치 설정
plt.show() # 그래프 출력
```
